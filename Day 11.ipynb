```python
import random

print("""
.------.            _     _            _    _            _    
|A_  _ |.          | |   | |          | |  (_)          | |   
|( \/ ).-----.     | |__ | | __ _  ___| | ___  __ _  ___| | __
| \  /|K /\  |     | '_ \| |/ _` |/ __| |/ / |/ _` |/ __| |/ /
|  \/ | /  \ |     | |_) | | (_| | (__|   <| | (_| | (__|   < 
`-----| \  / |     |_.__/|_|\__,_|\___|_|\_\ |\__,_|\___|_|\_\\
      |  \/ K|                            _/ |                
      `------'                           |__/           
""")


spades = [11, 2, 3, 4, 5, 6, 7, 8, 9, 10, 10, 10, 10]
diamonds = [11, 2, 3, 4, 5, 6, 7, 8, 9, 10, 10, 10, 10]
clubs = [11, 2, 3, 4, 5, 6, 7, 8, 9, 10, 10, 10, 10]
hearts = [11, 2, 3, 4, 5, 6, 7, 8, 9, 10, 10, 10, 10]

suits = [spades, diamonds, clubs, hearts]

def deal_card():
    """Returns a random card from the deck and removes that card from being able to be drawn again."""
    suit = random.choice(suits)
    card_index = random.randint(0,len(suit)-1)
    card = suit.pop(card_index)
    return card

def calculate_score(card_list):
    """Takes a list of cards and returns a score based on the sum of the cards"""
    score = sum(card_list)
    if score == 21 and len(card_list) == 2:
        # if blackjack, return score as 0
        return 0
    elif score > 21 and 11 in card_list:
        #elif score is over 21 and there is an ace, make ace value 1
        ace_index = card_list.index(11)
        card_list[ace_index] = 1
        new_score = calculate_score(card_list)
        return new_score
    else:
        return score
    
def compare(score1, score2):
    """Takes user and dealer scores as an input and compares them to determine a winner"""
    if score2 == 0:
        return "Dealer got a blackjack, you lose."
    elif score1 == 0:
        return "You got a blackjack, you win!"
    elif score1 > 21:
        return "You got over 21. Bust. You lose."
    elif score2 > 21:
        return "Dealer got over 21. Bust. You win!"
    elif score1 > score2:
        return "You beat the dealer!"
    elif score2 > score1:
        return "You lost to the dealer."
    else:
        return "It's a draw!"

user_cards = []
dealer_cards = []
game_over = False

for i in range(2):
    user_cards.append(deal_card())
    dealer_cards.append(deal_card())

while not game_over:
    user_score = calculate_score(user_cards)
    dealer_score = calculate_score(dealer_cards)
    
    print(f"\nUser cards {user_cards}, user score {user_score}")
    print(f"Dealer first card [{dealer_cards[0]}]\n")
    
    if user_score == 0:
        game_over = True
    elif dealer_score == 0:
        game_over = True
    elif user_score > 21:
        game_over = True
    else:
        another_card = input("Do you want to deal another card? 'Y' or 'N': ").upper()
        if another_card == 'Y':
            user_cards.append(deal_card())
        else:
            game_over = True

while dealer_score < 17 and dealer_score != 0:
    dealer_cards.append(deal_card())
    dealer_score = calculate_score(dealer_cards)
    
print(f"\nFinal user hand: {user_cards}, user score: {user_score}")
if user_score <= 21 and user_score != 0:
    print(f"Final dealer hand: {dealer_cards}, dealer score: {dealer_score}")
print(compare(score1 = user_score, score2 = dealer_score))
```
